### Criterio de fonación

Voy a decir que la señal está fonando si la distancia entre picos consecutivos es mayor a la distancia enre picos mínima de oscilación. Si es menor, digo que es ruido. 

Primero estudio las distancias entre picos.

```python
%matplotlib qt
plt.plot(t_,s_m_, label = 'sonido')
plt.plot(t_[picos], s_m_[picos], '.')
```

```python
%matplotlib inline
fig, axes = plt.subplots(6, 2, figsize=(15,22), constrained_layout=True, sharey=False)
axs = axes.flatten()


for i,ps in enumerate(pulsos):
    t_ = pulsos[ps]['t']
    s_m_ = pulsos[ps]['s_m']
    b_ = pulsos[ps]['b']
    picos, _ = signal.find_peaks(s_m_)  

    #guardo los indices de los picos
    pulsos[ps]['picos_m'] = picos

    #umbral distancia
    umbral = 1.5 * 1e-4
    umbral2 = 5* 1e-6
    #armo el array de distancias entr e picos consecutivos
    dist_picos = np.diff(t_[picos])
    prom_dist_picos = []
    
    index_fon = []
    for k in range(1,len(picos) -1):
        d_prom =  (t_[picos[k + 1]] - t_[picos[k - 1]]) / 2
        prom_dist_picos.append(d_prom)
        if d_prom > umbral:
            index_fon.append(picos[k])
    
    #print(len(picos_m), len(dist_picos))
    prom_dist_picos = np.array(prom_dist_picos)
    index_fon = np.array(index_fon)
    dif_dif = np.diff(prom_dist_picos)
    index_dif_dif_fon = picos[:-3][np.abs(dif_dif)<umbral2]
    
    axs[i].plot(t_[picos][:-1],dist_picos*1e4, label = 'Distancia a su pico consecutivo')
    axs[i].plot(t_[picos][1:-1],prom_dist_picos*1e4, label = 'Distancia promedio a su pico consecutivo', color = 'black')
    axs[i].axhline( umbral*1e4,label = f'umbral ', color = 'red', ls = '--')
    #axs[i].plot(t_[picos][1:-2],np.abs(dif_dif)*1e4, label = 'diferencia de dist')
    axs[i].plot(t_,s_m_, label = 'sonido')
    axs[i].plot(t_[index_fon],s_m_[index_fon], '.', label = 'fonacion')
    #axs[i].plot(t_[index_dif_dif_fon],s_m_[index_dif_dif_fon], '.', label = 'fonacion promedio')
    axs[i].set_title(f'Pulso {i+1}')
    axs[i].legend()
plt.show()
```

```python
%matplotlib inline
delta = np.zeros(len(t))
tiempos_fonacion = {ps: {'t_i': [] , 't_f': [] , 't_fon': pulsos[ps]['int_fon'][0] } for ps in pulsos}
for ps in pulsos:
    t_ = pulsos[ps]['t']
    b_ = pulsos[ps]['b']
    picos = pulsos[ps]['picos_m']

    #tiempo inicial de fonacion
    t_fon_i = pulsos[ps]['int_fon'][0]

    #indice de finaion inicial
    index_fon_i = np.where(t > t_fon_i)

    #indice de inicio de pulso
    index_i_pulso = np.where(t==t_[0])
    delta[index_i_pulso[0][0] : index_fon_i[0][0]] = 1

    #dict tiempos fonacion
    tiempos_fonacion[ps]['t_i'] = t_[0]
    tiempos_fonacion[ps]['t_f'] = t_[-1]
        
#delta[b<0] = 0
plt.figure(figsize = (15,4))
plt.xlabel('Tiempo [s]')
plt.plot(t,s_m/max(s_m), label = 'sonido')
plt.plot(t,delta, label = r'$\delta$')

plt.xlim(0.35, 0.7)
plt.legend()
plt.show()
```

```python
%matplotlib inline
delta1 = np.zeros(len(t))
delta2 = np.zeros(len(t))
t_fonacion = {ps: {'t_i': [] , 't_f': [] } for ps in pulsos}
for ps in pulsos:
    t_ = pulsos[ps]['t']
    b_ = pulsos[ps]['b']
    picos = pulsos[ps]['picos_m']

    #tiempo inicial de fonacion
    ti_fon = pulsos[ps]['int_fon'][0]
    tf_fon = pulsos[ps]['int_fon'][1]
    
    #tiempo inicial de la silaba
    ti_sil = t_[0]
    tf_sil = t_[-1]
    
    #construyo el delta
    delta1[(t >= ti_sil ) & (t <= ti_fon )] = 1
    delta2[(t >= ti_sil ) & (t <= ti_fon )] = 1
    delta2[(t >= tf_fon ) & (t <= tf_sil )] = 1

    t_fonacion[ps]['t_i'] = ti_fon
    t_fonacion[ps]['t_f'] = tf_fon
#delta[b<0] = 0
plt.figure(figsize = (15,4))
plt.xlabel('Tiempo [s]')
plt.plot(t,s_m/max(s_m), label = 'sonido')
plt.plot(t,delta1, label = r'$\delta_1$', alpha = .9)
plt.plot(t,delta2, label = r'$\delta_2$', alpha = .6)

plt.xlim(0.35, 0.7)
plt.legend()
plt.show()
```

```python
%matplotlib inline
fig, axes = plt.subplots(6, 2, figsize=(15,22), constrained_layout=True, sharey=False)
axs = axes.flatten()

pesos_umbral = [1, 1, 1.35, 1.2, 1.1, 1.1, 1.35, 1.1, .8, .8, .9, .7]
for i,ps in enumerate(pulsos):
    t_ = pulsos[ps]['t']
    s_m_ = pulsos[ps]['s_m']
    b_ = pulsos[ps]['b']
    picos_m, _ = signal.find_peaks(s_m_, height = 0 )  

    #guardo los indices de los picos
    pulsos[ps]['picos_m'] = picos_m
    
    #umbral de fonacion
    umbral_m = max(s_m_[picos_m]) * 0.3 * pesos_umbral[i] #sigma_m * 20

    #indices de los picos que pasan el umbral
    index_fon_m = [p for p in picos_m if s_m_[p] > umbral_m]

    #fraccion de los picos que pasan el umbral
    fon_m = len(index_fon_m)/ len(picos_m)

    #indices de los intervalos de fonacion (criterio: que no sean consecutivos)
    index_intervalo_fon_m = [index_fon_m[0]]
    tresh = max(np.diff(picos_m)) * 2 # criterio para determnar si los picos son consecutivos o no
    for j in range(len(index_fon_m)):
        if index_fon_m[j] - index_fon_m[j-1] > tresh:
            index_intervalo_fon_m.append(index_fon_m[j])
        if j < len(index_fon_m)-1:
            if index_fon_m[j+1] - index_fon_m[j] > tresh:
                index_intervalo_fon_m.append(index_fon_m[j])
    index_intervalo_fon_m.append(index_fon_m[-1])
    
    #guardo los indices de los intervalos de fonacion        
    pulsos[ps]['intervalos_fon_m'] = index_intervalo_fon_m

    axs[i].plot(t_,s_m_, label = 'sonido', color = 'orange')
    axs[i].axhline( umbral_m,label = f'umbral sonido pasa {np.round(fon_m*100)}%', color = 'red', ls = '--')
    axs[i].plot(t_[picos_m], s_m_[picos_m], '.', color = 'red')
    axs[i].plot(t_[index_fon_m], s_m_[index_fon_m], '.', color = 'darkorange')
    axs[i].plot(t_[index_intervalo_fon_m] , s_m_[index_intervalo_fon_m] , 'o', color = 'green')
    axs[i].plot(t_,b_, label = r'$\beta$')
    axs[i].set_title(f'Pulso {i+1}')
    axs[i].legend()
```
