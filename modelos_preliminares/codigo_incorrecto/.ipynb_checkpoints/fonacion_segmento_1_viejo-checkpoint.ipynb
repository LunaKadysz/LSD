### Umbral: porcentaje de la modulación
Primero calculamos todos los puntos que pasan ese umbral. Luego calculamos el porcentaje de esa señal que se enceuntra dentro del pulso de la presión

```python
peaks_bs, _ = signal.find_peaks(b, height = 0.4, distance = 500 ) 
plt.plot(t, b)
plt.plot(t[peaks_bs], b[peaks_bs], '.')
plt.title('Beta utilizado')

fracs = []
for i,p in enumerate(peaks_bs):
    if i%2==0:
        plt.plot(t[p],b[p], 'o', color = 'red')
        frac = ( b[peaks_bs[i]]) / (b[peaks_bs[i+1]])
        fracs.append(frac)
        print(f'fracción con su pico posterior: {frac}')
umbral = np.mean(fracs) # 1 - mean(fracs). Igualmente: conviene sacar un promedio? no conviente para cada puslo tener un umbral distinto? CHARLAR
print(f'promedio: {umbral}')
```

#### Segunda aproximación: 
el umbral va a estar determinado por la modulación del pulso ($\beta$). Tomo como umbral el 50% de la señal modulada por la fracción de $\beta$

```python
index_split = []
for i in range(len(t)-1):
    if b[i]*b[i+1] < 0:
        #print(f'Cambio de signo en t= {t[i]}')
        index_split.append(i)
print(len(index_split))
print(index_split)
index_split.pop(0)
print(index_split)

plt.figure(figsize=(14,4))
plt.plot(t,s_s)
plt.plot(t,b)
plt.plot(t[index_split], b[index_split], 'o')
```

```python
#ordeno los datos

pulsos = {i: {'t' : t[index_split[i]:index_split[i+1]],
              's_s': s_s[index_split[i]:index_split[i+1]],
              's_m': s_m[index_split[i]:index_split[i+1]],
              'b': b[index_split[i]:index_split[i+1]]} for i in range(24) if i%2==0 } #ME FALTA EL ULTIMO PULSO
print(len(pulsos))
```

```python
fig, axes = plt.subplots(6, 2, figsize=(15,20), constrained_layout=True, sharey=False)
axs = axes.flatten()

for i,ps in enumerate(pulsos):
    t_ = pulsos[ps]['t']
    s_s_ = pulsos[ps]['s_s']
    s_m_ = pulsos[ps]['s_m']
    b_ = pulsos[ps]['b']
    umbral_s = b_ * max(s_s_) * 0.5
    umbral_m = b_ * max(s_m_) * 0.5
    axs[i].plot(t_,s_s_, label = 'simulacion')
    axs[i].plot(t_,s_m_, label = 'sonido')
    axs[i].plot(t_, umbral_s, label = 'umbral sim', color = 'blue', ls = '--')
    axs[i].plot(t_, umbral_m, label = 'umbral sonido', color = 'orange', ls = '--')
    axs[i].plot(t_,b_, label = r'$\beta$')
    axs[i].set_title(f'Pulso {i+1}')
    axs[i].legend()
    
```

```python
fig, axes = plt.subplots(6, 2, figsize=(15,20), constrained_layout=True, sharey=False)
axs = axes.flatten()

for i,ps in enumerate(pulsos):
    t_ = pulsos[ps]['t']
    s_s_ = pulsos[ps]['s_s']
    s_m_ = pulsos[ps]['s_m']
    b_ = pulsos[ps]['b']
    picos_s, _ = signal.find_peaks(s_s_, height = 0 )  
    picos_m, _ = signal.find_peaks(s_m_, height = 0 )  

    umbral_s = b_ * max(s_s_) * 0.5
    umbral_m = b_ * max(s_m_) * 0.5

    fon_s = len([ps for ps in picos_s if s_s_[ps] > umbral_s[ps]])/ len(picos_s)
    fon_m = len([pm for pm in picos_m if s_m_[pm] > umbral_m[pm]])/ len(picos_m)

    #print(f'Porcentaje de fonación para el pulso {i+1}:')
    #print(f'Simulación: {fon_s * 100}%')
    #print(f'Medición: {fon_m * 100}%')
    axs[i].plot(t_,s_s_, label = 'simulacion')
    axs[i].plot(t_,s_m_, label = 'sonido')
    axs[i].plot(t_, umbral_s, label = f'umbral sim: pasa {np.round(fon_s*100)}%', color = 'blue', ls = '--')
    axs[i].plot(t_, umbral_m, label = f'umbral sonido pasa {np.round(fon_m*100)}%', color = 'orange', ls = '--')
    axs[i].plot(t_[picos_s], s_s_[picos_s], '.', color = 'blue')
    axs[i].plot(t_[picos_m], s_m_[picos_m], '.', color = 'orange')
    axs[i].plot(t_,b_, label = r'$\beta$')
    axs[i].set_title(f'Pulso {i+1}')
    axs[i].legend()
```
