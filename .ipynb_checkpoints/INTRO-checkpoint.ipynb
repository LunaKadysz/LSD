## Dinámica del órgano vocal del Chingolo

La idea de esta notebook es jugar un poco con las ecuaciones del modelo que describe la dinámica del órgano vocal del Chingolo .
Esta práctica está basada en este paper: https://www.researchgate.net/publication/12917420_The_neuromuscular_control_of_birdsong

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import seaborn as sns

sns.set_theme()
sns.set_style("darkgrid")
```

La ecuación que modela la dinámica del organo vocal de un ave:
$$
\ddot{x} + (x^2 - \beta) \dot{x} + k x - \delta = 0
$$

donde 
$\delta$ es la fuerza externa.

$(x^2 - \beta) \dot{x}$ el término disipativo. Compite la entrega de energía $-\beta \dot{x}$ y la disippación no lineal $x^2\dot{x}$.



podemos reescribirla en términos de la derivada

$$
\frac{d}{dt}\big( \dot{x} + \mu (\frac{x^3}{3} - \beta x)\big) = - k(x + \delta/k)
$$

si definimos $ \omega =  \dot{x} + \mu (\frac{x^3}{3} - \beta x) $ nos termina quedando el siguiente sistema de ecuaciones

$$
\begin{cases}
   \frac{d \omega}{dt} = - k(x + \delta/k)\\
    \frac{d x}{dt} = \omega - \mu ( \frac{x^3}{3} - \beta x)  \\
  \end{cases}
$$

absorbo la $\mu$ dentro de la nueva variable $\tilde{\omega}$
Finalmente el sistema de ecuaciones nos queda

$$
\begin{cases}
   \frac{d \tilde{\omega}}{dt} = 10000 \frac{k}{\mu} (x + \frac{\delta}{k})\\
    \frac{d x}{dt} = 10000 \mu (\omega - \frac{ c x^3}{\mu} - \frac{\beta x}{\mu})  \\
  \end{cases}
$$

Creo que hay un signo mal, me fijé en el paper y las ecuaciones quedan


$$
\begin{cases}
   \frac{d \tilde{\omega}}{dt} = 10000 \frac{k}{\mu} (x + \frac{\delta}{k})\\
    \frac{d x}{dt} = 10000 \mu (\omega - \frac{ c x^3}{\mu} + \frac{\beta x}{\mu})  \\
  \end{cases}
$$


#### Comentario
La solución me está divergiendo. Mi primera hipótesis es que tiene que ver con el paso de integración.


Para encontrar un dt de integración adecuado voy a calcular la escala característica del problema.
Puedo pedir la condición CFL (eso era para derivadas parciales) o los autovalores de la matriz Jacobiana del sistema. Donde se cumple que $T_{caracterisitco} = \frac{1}{|\lambda_{max}|}$

$$
\begin{cases}
   \dot{\tilde{\omega}} = 10000 \frac{k}{\mu} (x + \frac{\delta}{k})\\
    \dot{x} = 10000 \mu (\omega - \frac{ c x^3}{\mu} - \frac{\beta x}{\mu})  \\
  \end{cases} 	\Longrightarrow
  J = \begin{pmatrix}0 & 10^4 \frac{k}{\mu} \\ 10^4 \mu & 10^4 \beta \end{pmatrix} 
$$

Busco los autovals de J:

$$
\det(J - \lambda I) = 0 = 
\begin{vmatrix} 
-\lambda & 10^4 \frac{k}{\mu} \\ 
10000 \mu & 10^4 \beta - \lambda
\end{vmatrix} = 0
$$

$$
(-\lambda)(10^4 \beta - \lambda) - (10^4 \frac{k}{\mu})(10^4 \mu) = 0
\\
\lambda^2 - 10^4 \beta \lambda + 10^8 k = 0
\\
\lambda_{\pm} = \frac{10^4 \beta \pm \sqrt{(10^4 \beta)^2 - 4(10^8 k)}}{2}
$$
Luego, tenemos a los autovalores
$
\lambda_{\pm} = 5000 (\beta \pm \sqrt{\beta^2 - 4k})
$.
Por ende
$$
\lambda_{max} = 5000(|\beta| + \sqrt{|\beta^2 - 4k|})
\Longrightarrow T_{caracteristico} = \frac{1}{|\lambda_{max}|} = \frac{1}{5000(|\beta| + \sqrt{|\beta^2 - 4k|})}
$$

Como quiero garantizar estabilidad numérica, pido $dt << T_{caracteristico}$.Bajo un orden (o dos?):
$$
T_{caracteristico} = \frac{1}{|\lambda_{max}|} = \frac{1}{5000(|\beta| + \sqrt{|\beta^2 - 4k|})} > \frac{1}{10^4(|\beta| + \sqrt{|\beta^2 - 4k|})}  > \frac{1}{10^5(|\beta| + \sqrt{|\beta^2 - 4k|})} = \frac{1}{20 * |\lambda_{max}|} \sim dt
$$

El factor ese 10^4 lo que hace es reducir el tiempo caracteristico del sistema. Con este factor, el sistema va a evolucionar $10^4$ veces más rápido que las originales.

Veamos un poco los números que me pasó Gabo:

$$
k=1, c=1 , \mu \sim 10^5
$$


```python
# calculo lambda max con los valores de los parámetros que me dio gabo
k = 1
mu = 1e5
beta = 1  # a ver

#calculo los lambdas
lambda1 = 5000 * (beta + np.sqrt(abs(beta**2 - 4*k)))
lambda2 = 5000 * (beta - np.sqrt(abs(beta**2 - 4*k)))
lambda_max = max(abs(lambda1), abs(lambda2))

# calculo el dt
dt = 1 / (20 * lambda_max)

print(r"  λ1 = ",lambda1 , r"\n λ2 =", lambda2)
print(f"Tc = {1/lambda_max}")
print(f" dt = {dt}")
```

### Integracion numérica

```python


#esta f es la funcion en donde escribo el sistema de funciones
def f(X,t,params):
    w, x = X
    k, mu, delta, c, beta = params
    
    f_w = 1e4 * k / mu * (x + delta / k) # el factor 1e4 era para acomodar las unidades en s?
    f_x = 1e4 * mu * (w - c * x**3 / mu + beta * x / mu )
    # f_x = 1e4 * (mu * w + beta * x ) # probé sacarle el término no lineal
    return np.array([f_w,f_x])


def paso_RK(X, f, t, dt, params, orden = 2):
    k1 = f(X,t,params)
    k2 = f(X+k1*dt/2,t,params)

    if orden == 2: #el default es orden 2 a no ser que aclare
        X = X + (k1 + k2)/2*dt
    elif orden == 4:
        k3 = f(X+k2*dt/2,t,params)
        k4 = f(X+k3*dt,t,params)
        X = X + (k1 + 2*(k2+k3) + k4)/6*dt
    #t = t + dt
    return X
```

## Modelo exacto del paper

Definiendo la ecuación como un sistema de primer orden para dos variables $u,v$ queda:

$$ 
  \begin{cases}
  \dot{x} = y \\
    \dot{y} = -kx -cx^2 y + \beta y - \delta \\
  \end{cases}chic
$$
utilizando unidades de dyn, s, cm, g:
$c=10^8 , \delta = 0 , k=10^9 $. Con $ \beta = 1000$ debería ver oscilaciones del orden de los 5kHz $\Rightarrow T_rapido = 0.0002s$

```python
1/5000
```

```python

#funcion del sistema de ecuaciones
def f(X, t, params):
    x,y = X
    k, c, delta, beta = params
    
    f_x = y
    f_y = -k*x - c * x**2 * y + beta * y - delta
    return np.array([f_x,f_y])
    
#método Runge-Kutta 4
def paso_RK(X, f, t, dt, params, orden=4):
    k1 = f(X, t, params)
    k2 = f(X + k1 * dt / 2, t, params)
    
    if orden == 2:
        X = X + (k1 + k2) / 2 * dt
    elif orden == 4:
        k3 = f(X + k2 * dt / 2, t, params)
        k4 = f(X + k3 * dt, t, params)
        X = X + (k1 + 2 * (k2 + k3) + k4) / 6 * dt
    return X

```

### Caso $\delta (t) = 0 $, $\beta (t) = cte$

```python

tf = 0.01
dt = 1e-5  
N = int(tf / dt)

#basados en el paper
k, c = 1e9 , 1e8
delta = np.zeros(N + 1)
beta = np.ones(N + 1) * (1e3)  


x0 = 1
#w0 = 0.01 # le doy una perturbacion chica
w0 = 0
X0 = np.array([w0, x0])

ts = np.arange(0, N + 1) * dt
Xs = np.zeros((N + 1, 2))
Xs[0] = X0

```

#### Integro para distintos valores de $\beta$
$c=10^8 , \delta = 0 , k=10^9 $

```python
#voy a integrar para distitnos valores de beta

betas = [-1000, 0, 1000, 3000]

fig, [ax1, ax2,ax3] = plt.subplots(1, 3, figsize=(15,5), constrained_layout=True,
                                sharey=False)
colors = ['green','red','black',"blue"]

for i,b in enumerate(betas):
    Xs = np.zeros((N + 1, 2))
    Xs[0] = X0
    # Integración
    for j in range(N):
        params = [k, c, delta[j], b]
        Xs[j + 1] = paso_RK(Xs[j], f, ts[j], dt, params, 4)
    
    
    #ploteo
    ax1.plot(ts,Xs[:,0],alpha = .60, color=colors[i], label=  r"$\beta$ = %s"%b)
    ax2.plot(ts,Xs[:,1],alpha = .60, color=colors[i], label=  r"$\beta$ = %s"%b )
    ax3.plot(Xs[:,0],Xs[:,1],color = colors[i], label=  r"$\beta$ = %s"%b)
    
    ax1.set_xlabel('$Tiempo$', fontsize=14)
    ax1.set_ylabel('$x$', fontsize=14)
    ax1.set_title('x (t)', fontsize=16)
    ax2.set_xlabel('$Tiempo$', fontsize=14)
    ax2.set_ylabel(r'$y = \dot{x}$', fontsize=14)
    ax2.set_title(r'$y(t) = \dot{x}(t)$', fontsize=16)
    ax3.set_title("Espacio de fases",fontsize=16)
    ax3.set_ylabel("$x$", fontsize=14)
    ax3.set_xlabel("$w$",fontsize=14)
ax1.legend()
ax2.legend()
ax3.legend()
fig.suptitle(r"Evolución del sistema para distintos valores de $\beta$")
```

Nos da lo que esperábamos del paper: si $\beta < 0$ no vemos oscilaciones.

#### Integro para distintos valores de k
$c=10^8 , \delta = 0 , \beta = 1000$

```python
#voy a integrar para distitnos valores de k

ks = [4e8, 8e8, 1.2e9]

fig, [ax1, ax2,ax3] = plt.subplots(1, 3, figsize=(15,5), constrained_layout=True,
                                sharey=False)
colors = ['green','red','black',"blue"]

for i,k_ in enumerate(ks):
    Xs = np.zeros((N + 1, 2))
    Xs[0] = X0
    # Integración
    for j in range(N):
        params = [k_, c, delta[j], beta[j]]
        Xs[j + 1] = paso_RK(Xs[j], f, ts[j], dt, params, 4)
    
    
    #ploteo
    ax1.plot(ts,Xs[:,0],alpha = .60, color=colors[i], label=  f"k = {k_}")
    ax2.plot(ts,Xs[:,1],alpha = .60, color=colors[i], label=  f"k = {k_}" )
    ax3.plot(Xs[:,0],Xs[:,1],color = colors[i], label=  f"k = {k_}")
    
    ax1.set_xlabel('$Tiempo$', fontsize=14)
    ax1.set_ylabel('$x$', fontsize=14)
    ax1.set_title('x (t)', fontsize=16)
    ax2.set_xlabel('$Tiempo$', fontsize=14)
    ax2.set_ylabel(r'$y = \dot{x}$', fontsize=14)
    ax2.set_title(r'$y(t) = \dot{x}(t)$', fontsize=16)
    ax3.set_title("Espacio de fases",fontsize=16)
    ax3.set_ylabel("$x$", fontsize=14)
    ax3.set_xlabel("$w$",fontsize=14)
ax1.legend()
ax2.legend()
ax3.legend()
fig.suptitle(r"Evolución del sistema para distintos valores de $\beta$")
```
